```python
# Import the necessary libraries

import textract
import string
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Extract text from a PDF document using textract.

text = textract.process('/content/The_Adventures_of_Tom_Sawyer_-_Penguin_Readers-min.pdf').decode('utf-8')
```

```python
text
```

```python
len(text)
```

```python
text = text.replace("\n", " ").replace("\x0c", " ")
```

```python
text
```

```python
# Remove punctuation from the 'text' variable and store the result in 'text_punc_removed'.

text_punc_removed =''
for word in text:
  if word not in string.punctuation:
    text_punc_removed = text_punc_removed+word
```

```python
# Convert the text to lowercase

text_punc_removed=text_punc_removed.lower()
```

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
# Tokenize the text to split it into individual words

word = nltk.word_tokenize(text_punc_removed)
```

```python
word
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
# Get a list of English stop words from the nltk corpus.

stopwords=nltk.corpus.stopwords.words('english')
```

```python
# Remove stop words from the list of words.

filtered_words = [i for i in word if i not in stopwords]
```

```python
# Combine the filtered words into a single string.

filtered_text = ' '.join(filtered_words)
```

```python
filtered_text
```

```python
# Create a word cloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)
```

```python
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
